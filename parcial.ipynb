```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import tabulate

df = pd.read_csv('Datos.csv')
print(df.head().to_markdown(index=False, stralign='center', numalign='center', tablefmt='heavy_grid'))
```

seleccionar una columna para agrupar los datos y aplicar algún proceso sobre esta agrupación (sumar ventas, contar ocurrencias, etc.).

```python
pago_salario = df.groupby('Cargo')['Salario'].sum()
```

creacion de graficos

```python
plt.figure (figsize=(17,4))
plt.bar(pago_salario.index, pago_salario.values)
plt.xlabel('Empleado')
plt.ylabel('cargo')
plt.title('Gastos de la empresa en salarios')
plt.xticks(rotation=45)
plt.show()
```

```python
df['Fecha_Ingreso'] = pd.to_datetime(df['Fecha_Ingreso'])
pagos_fecha = df.groupby('Fecha_Ingreso')['Salario'].sum()
```

```python
plt.figure (figsize=(6,4))
plt.plot(pagos_fecha.index, pagos_fecha.values)
plt.xlabel('ventas_total')  
plt.ylabel('Fecha_Ingreso')
plt.title('pagos de la empresa en salarios')
plt.show()
```

calcular la media, mediana, moda, desviación estándar y varianza de alguna de las columnas numéricas del dataset utilizando NumPy.


```python
media = np.mean(pago_salario)
print('La media de los salarios es : ', media)

mediana = np.median(pago_salario)
print('La mediana de los salarios es : ', mediana)

moda = stats.mode(pago_salario)
print('La moda de los salarios es : ', moda.mode, ' y este se repite '
, moda.count, ' veces')

desviacion_estandar = np.std(pago_salario)
print('La desviacion estandar de los salarios es : ', desviacion_estandar)

varianza = np.var(pago_salario)
print('La varianza de los salarios es : ', varianza)

```
