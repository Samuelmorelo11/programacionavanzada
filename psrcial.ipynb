```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt

import tabulate

df = pd.read_csv('Datos.csv')
print(df.head().to_markdown(index=False, stralign='center', numalign='center', tablefmt='heavy_grid'))

```

seleccionar una columna para agrupar los datos y aplicar algún proceso sobre esta agrupación (sumar ventas, contar ocurrencias, etc.).

```python
col_salario = df['Salario']
sumatoria = np.sum(col_salario)
print('La sumatoria de los gastos de la empresa en salarios de los empleados es igual a: ', sumatoria)
```

creacion de graficos

```python
plt.figure (figsize=(4,2))
plt.bar(col_salario.index, col_salario.values)
plt.xlabel('Empleado')
plt.ylabel('Salario')
plt.title('Gastos de la empresa en salarios')
plt.xticks(rotation=45)
plt.show()
```

```python
ventas_fecha = df['Fecha_Ingreso']
sumatoria = np.sum(ventas_fecha)
```

calcular la media, mediana, moda, desviación estándar y varianza de alguna de las columnas numéricas del dataset utilizando NumPy.


```python
media = np.mean(col_salario)
print('La media de los salarios es : ', media)

mediana = np.median(col_salario)
print('La mediana de los salarios es : ', mediana)

moda = stats.mode(col_salario)
print('La moda de los salarios es : ', moda.mode, ' y este se repite '
, moda.count, ' veces')

desviacion_estandar = np.std(col_salario)
print('La desviacion estandar de los salarios es : ', desviacion_estandar)

varianza = np.var(col_salario)
print('La varianza de los salarios es : ', varianza)

```
